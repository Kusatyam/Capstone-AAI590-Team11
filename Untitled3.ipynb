### Load the IMU file

```python
import pandas as pd

# Load IMU data
imu_filtered_df = pd.read_csv(
    "/content/imu_filtered.txt",        # adjust path if different
    comment="#",                    # ignore header comment
    delim_whitespace=True,          # whitespace separator
    names=["timestamp", "wx", "wy", "wz", "ax", "ay", "az"]  # column names
)

print("Shape:", imu_filtered_df.shape)
display(imu_filtered_df.head())
```

### Convert timestamps to readable datetime

```python
imu_filtered_df["timestamp"] = pd.to_datetime(imu_filtered_df["timestamp"], unit="ns")
display(imu_filtered_df.head())
```

### Basic data inspection

```python
# Info and summary stats
print(imu_filtered_df.info())
display(imu_filtered_df.describe().T)

# Check for missing values and duplicates
print("\nMissing values per column:")
print(imu_filtered_df.isna().sum())

print("\nDuplicates:", imu_filtered_df.duplicated().sum())
```

### find missing or NaN values (should be very few):

```python
imu_filtered_df = imu_filtered_df.ffill()
```

```python
# Handle missing values and duplicates
imu_filtered_df = imu_filtered_df.ffill()          # forward fill missing data
imu_filtered_df = imu_filtered_df.drop_duplicates() # remove duplicate rows
```

### Plots

```python
import matplotlib.pyplot as plt

# Plot angular velocity
imu_filtered_df.plot(x="timestamp", y=["wx", "wy", "wz"], title="Angular Velocity", figsize=(10,4))
plt.show()

# Plot linear acceleration
imu_filtered_df.plot(x="timestamp", y=["ax", "ay", "az"], title="Linear Acceleration", figsize=(10,4))
plt.show()

# Correlation heatmap
import seaborn as sns
sns.heatmap(imu_filtered_df[["wx","wy","wz","ax","ay","az"]].corr(), annot=True, cmap="coolwarm")
plt.title("Sensor Correlations")
plt.show()
```

```python
import numpy as np, pandas as pd, matplotlib.pyplot as plt, seaborn as sns
plt.style.use("seaborn-v0_8-whitegrid")

# Basic overview
print("Rows:", len(imu_filtered_df))
print("Time range:", imu_filtered_df['timestamp'].min(), "", imu_filtered_df['timestamp'].max())

# Compute sampling interval and frequency
imu_filtered_df['dt'] = imu_filtered_df['timestamp'].diff().dt.total_seconds()
avg_dt = imu_filtered_df['dt'].mean()
print(f"\nAverage sampling interval: {avg_dt:.6f} s  (~{1/avg_dt:.2f} Hz)")
print(f"Missing timestamps (gaps > 3median dt): {(imu_filtered_df['dt']>3*imu_filtered_df['dt'].median()).sum()}")
```

### Statistical Summary

```python
summary = imu_filtered_df[['wx','wy','wz','ax','ay','az']].describe().T
summary['variance'] = imu_filtered_df[['wx','wy','wz','ax','ay','az']].var().values
summary['skew']     = imu_filtered_df[['wx','wy','wz','ax','ay','az']].skew().values
summary['kurtosis'] = imu_filtered_df[['wx','wy','wz','ax','ay','az']].kurtosis().values
display(summary.round(6))
```

### Distributions and Outliers

```python
fig, axes = plt.subplots(2,3,figsize=(14,8))
cols = ['wx','wy','wz','ax','ay','az']
for i,c in enumerate(cols):
    sns.histplot(imu_filtered_df[c], kde=True, ax=axes[i//3][i%3])
    axes[i//3][i%3].set_title(f"Distribution of {c}")
plt.tight_layout()
plt.show()

# Boxplots for outliers
plt.figure(figsize=(10,4))
sns.boxplot(data=imu_filtered_df[cols])
plt.title("Outlier Detection")
plt.show()
```

### Correlation and Cross-Axes Relationships

```python
corr = imu_filtered_df[['wx','wy','wz','ax','ay','az']].corr()
plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Matrix")
plt.show()

# Pairwise scatter
sns.pairplot(imu_filtered_df.sample(500)[['wx','wy','wz','ax','ay','az']])
plt.suptitle("Pairwise Relationships (500-sample subset)", y=1.02)
plt.show()
```

### Drift and Bias Trend Analysis

```python
imu_filtered_df['cum_ax'] = imu_filtered_df['ax'].cumsum()
imu_filtered_df['cum_ay'] = imu_filtered_df['ay'].cumsum()
imu_filtered_df['cum_az'] = imu_filtered_df['az'].cumsum()

plt.figure(figsize=(10,4))
plt.plot(imu_filtered_df['timestamp'], imu_filtered_df[['cum_ax','cum_ay','cum_az']])
plt.title("Cumulative Acceleration (Drift Indicator)")
plt.ylabel(" a (units)")
plt.show()
```

Linear growth  systematic bias; oscillatory behavior  well-balanced sensors.

### Spectral (FFT) Analysis

```python
from scipy.fft import rfft, rfftfreq
import numpy as np

fs = 1/avg_dt
N = len(imu_filtered_df)
yf = np.abs(rfft(imu_filtered_df['ax'] - imu_filtered_df['ax'].mean()))
xf = rfftfreq(N, avg_dt)

plt.figure(figsize=(10,4))
plt.plot(xf, yf)
plt.xlim(0, fs/2)
plt.xlabel("Frequency (Hz)")
plt.ylabel("|FFT(ax)|")
plt.title("Dominant Frequencies in Acceleration")
plt.show()
```

vibration frequency peaks  useful for identifying mechanical resonance or control oscillations.

### Outlier Flagging (Z-Score)

```python
from scipy.stats import zscore
z = np.abs(zscore(imu_filtered_df[['wx','wy','wz','ax','ay','az']]))
outliers = (z>3).sum()
print("Outlier count per column:\n", outliers)
```

### Rolling Window Statistics (Sensor Stability Over Time)

```python
window = 200  # roughly 4 seconds if 50 Hz
imu_filtered_df["ax_roll_mean"] = imu_filtered_df["ax"].rolling(window).mean()
imu_filtered_df["ax_roll_std"]  = imu_filtered_df["ax"].rolling(window).std()

plt.figure(figsize=(12,5))
plt.plot(imu_filtered_df["timestamp"], imu_filtered_df["ax"], alpha=0.5, label="ax (raw)")
plt.plot(imu_filtered_df["timestamp"], imu_filtered_df["ax_roll_mean"], color='red', label="Rolling Mean (200)")
plt.fill_between(imu_filtered_df["timestamp"],
                 imu_filtered_df["ax_roll_mean"] - imu_filtered_df["ax_roll_std"],
                 imu_filtered_df["ax_roll_mean"] + imu_filtered_df["ax_roll_std"],
                 color='orange', alpha=0.2, label="1 std")
plt.legend(); plt.title("Rolling Statistics  Linear Acceleration X"); plt.show()
```

### Principal Component Analysis (PCA)

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

X = imu_filtered_df[["wx","wy","wz","ax","ay","az"]].dropna()
X_scaled = StandardScaler().fit_transform(X)

pca = PCA()
pca.fit(X_scaled)

explained = np.cumsum(pca.explained_variance_ratio_)
plt.figure(figsize=(6,4))
plt.plot(np.arange(1,7), explained, marker='o')
plt.title("Cumulative Variance Explained by PCA Components")
plt.xlabel("Number of Components")
plt.ylabel("Cumulative Variance")
plt.grid(True)
plt.show()

print("Component loadings:\n", pd.DataFrame(pca.components_, columns=X.columns))
```

### Segmentation by motion state

```python
imu_filtered_df["rot_speed"] = np.sqrt(imu_filtered_df["wx"]**2 + imu_filtered_df["wy"]**2 + imu_filtered_df["wz"]**2)
imu_filtered_df["motion_state"] = np.where(imu_filtered_df["rot_speed"] > imu_filtered_df["rot_speed"].median(), "active", "steady")

sns.boxplot(x="motion_state", y="ax", data=imu_filtered_df)
plt.title("Acceleration Distribution by Motion State")
plt.show()
```
